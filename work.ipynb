```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt 
import numpy as np
```

```python
mdf = pd.read_csv('linked_mortality_file_1999_2000.csv') # Load mortality file
print( mdf.head() )
gdf = pd.read_sas("DEMO.xpt", format="xport") # Load demographics file
print( gdf.head() )
df = gdf.merge(mdf, on="SEQN", how="inner") # Merge mortality and demographics on SEQN variable
```

```python
# merging mortality and caffeine data
cdf = pd.read_sas("SSCAFE_A.xpt", format="xport") # Load caffeine file
df_1 = cdf.merge(df, on="SEQN", how="inner") # Merge caffeine and the already merged demo + mort on SEQN variable
df_1.to_csv("merged_demo_mortality_caffeine.csv", index=False)
```
